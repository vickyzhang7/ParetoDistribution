```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto
import pandas as pd

def median_based_mad_upper_bound(k, alpha):
    return np.sqrt(alpha) * ((2 ** (1 / alpha)) - 1) * np.sqrt(alpha - 2) / k


k = np.linspace(1, 3, 100)

alpha = 3 
median_based_mad_bound = median_based_mad_upper_bound(k, alpha)


pareto_dist = pareto(alpha)
cdf = 1 - pareto_dist.cdf(k)

pareto_intersection_index = np.argmin(np.abs(cdf - median_based_mad_bound))
pareto_intersection_k = k[pareto_intersection_index]
pareto_intersection_prob = cdf[pareto_intersection_index]

plt.plot(k, cdf, label="Pareto Distribution CDF", linestyle='--', color='k')
plt.plot(k, median_based_mad_bound, label="MAD Upper Bound (about median)", linestyle='-', color='k')

plt.plot(pareto_intersection_k, pareto_intersection_prob, 'ko', label="Intersection Point (Pareto)")
plt.xlabel('k')
plt.ylabel('Probability')
plt.title("Pareto Distribution CDF vs. MAD-based Upper Bounds")
plt.legend()
plt.grid(True)
plt.show()

print("Intersection Point:")
print("k =", np.round(pareto_intersection_k, 4))
print("Probability =", np.round(pareto_intersection_prob, 4))
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pareto
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from scipy.optimize import newton
import matplotlib.pyplot as plt
# Load the data
df = pd.read_csv('Data9622.csv')  

columns = [col for col in df.columns if col.startswith("W")]

# Initialize lists to store results
alpha_values = {'Year': [], 'Fit': [], 'Alp(MLE)': [], 'Alp(MAD)': [],'RMSE(MLE)': [], 'RMSE(MAD)': [],'Resi(MLE)': [], 'Resi(MAD)': []}

# Number of iterations for each year
num_iterations = 100

# Iterate through the years
for year_column in columns:
    # Lists to store results for each iteration
    alpha_mle_values = []
    alpha_mad_values = []
    rmse_mle_values = []
    rmse_mad_values = []
    resi_mle_values = []
    resi_mad_values = []
    
    for _ in range(num_iterations):
        year_data = df[year_column].dropna()


        # Split data into training and testing sets
        train_data, test_data = train_test_split(year_data, test_size=0.5)
        
        #fit python         
        param = pareto.fit(train_data)[0]
        

        # Fit the data to the Pareto distribution and get the alpha parameters using MLE
        beta_MLE = np.min(train_data)
        alpha_MLE = len(train_data) / np.sum(np.log(train_data/beta_MLE))

        # Fitting Pareto distribution on the training set for MLE
        pareto_params_MLE = (alpha_MLE, beta_MLE)

        # Calculate residuals and RMSE on the testing set for MLE
        test_residuals_MLE = test_data - pareto(*pareto_params_MLE).rvs(len(test_data))
        test_rmse_MLE = np.sqrt(mean_squared_error(test_data, pareto(*pareto_params_MLE).rvs(len(test_data))))

        # Fit the data to the Pareto distribution and get the alpha parameters using MAD
        median = np.median(train_data)
        mad = np.mean(np.abs(train_data - median))
        beta_MAD = np.min(train_data)

        # Define the equation for MAD
        def equation(alpha):
            return mad - (alpha * beta_MAD * (np.power(2, 1/alpha) - 1) / (alpha - 1))

        # Solve the equation for alpha using Newton's method
        alpha_MAD = newton(equation, 1.1)

        # Fitting Pareto distribution on the training set for MAD
        pareto_params_MAD = (alpha_MAD, beta_MAD)

        # Calculate residuals and RMSE on the testing set for MAD
        test_residuals_MAD = test_data - pareto(*pareto_params_MAD).rvs(len(test_data))
        test_rmse_MAD = np.sqrt(mean_squared_error(test_data, pareto(*pareto_params_MAD).rvs(len(test_data))))

        # Append results to lists
        alpha_mle_values.append(alpha_MLE)
        alpha_mad_values.append(alpha_MAD)
        rmse_mle_values.append(test_rmse_MLE)
        rmse_mad_values.append(test_rmse_MAD)
        resi_mle_values.append(np.mean(test_residuals_MLE))
        resi_mad_values.append(np.mean(test_residuals_MAD))
    
    # Calculate averages for each metric
    avg_alpha_mle = np.mean(alpha_mle_values)
    avg_alpha_mad = np.mean(alpha_mad_values)
    avg_rmse_mle = np.mean(rmse_mle_values)
    avg_rmse_mad = np.mean(rmse_mad_values)
    avg_resi_mle = np.mean(resi_mle_values)
    avg_resi_mad = np.mean(resi_mad_values)

    # Add results to the dictionary
    alpha_values['Year'].append(int(year_column[1:]))
    alpha_values['Fit'].append(param)
    alpha_values['Alp(MLE)'].append(avg_alpha_mle)
    alpha_values['Alp(MAD)'].append(avg_alpha_mad)
    alpha_values['RMSE(MLE)'].append(avg_rmse_mle)
    alpha_values['RMSE(MAD)'].append(avg_rmse_mad)
    alpha_values['Resi(MLE)'].append(avg_resi_mle)
    alpha_values['Resi(MAD)'].append(avg_resi_mad)

# Create DataFrame from the dictionary
results_df = pd.DataFrame(alpha_values)

# Convert 'Year' column to integer
results_df['Year'] = results_df['Year'].astype(int)

# # Round the values to one decimal place except for the 'Year' column
# results_df.iloc[:, 1:] = results_df.iloc[:, 1:].round(1)

# # Print the results
# print(results_df)

# Create new columns for absolute relative differences between alpha values and 'Fit'
results_df['MLE % err'] = abs(results_df['Alp(MLE)'] - results_df['Fit']) / results_df['Fit']*100
results_df['MAD % err'] = abs(results_df['Alp(MAD)'] - results_df['Fit']) / results_df['Fit']*100

# Remove 'Resi(MLE)' and 'Resi(MAD)' columns
results_df.drop(columns=['Resi(MLE)', 'Resi(MAD)'], inplace=True)

# Re-round values after adding new columns and removing unwanted ones
results_df.iloc[:, 1:] = results_df.iloc[:, 1:].round(1)

# Print the updated DataFrame
print(results_df)


plt.figure(figsize=(10, 6))  # Adjusted size for readability
# plt.plot(results_df['Year'], results_df['Fit'], label='Alphy_Fit', linestyle='-', color='k', marker='o')  # Black line
plt.plot(results_df['Year'], results_df['Alp(MLE)'], label='α(MLE)', linestyle='--', color='k', marker='o')  # Black line
plt.plot(results_df['Year'], results_df['Alp(MAD)'], label='α(MAD)', linestyle='-', color='k', marker='*')  # Black line

# Adding axis labels, legend, grid, and title
plt.xlabel('Year', fontsize=14)  # Increased font size
plt.ylabel('α', fontsize=14)  # Increased font size
plt.title('Alp(MLE), and Alp(MAD) Over Years', fontsize=16)  # Larger title
plt.legend(loc='upper left', fontsize=12)  # Legend in the upper right corner with larger font

# Rotate x-axis labels to avoid crowding
plt.xticks(results_df['Year'], rotation=90, fontsize=12)  # Rotate labels vertically, ensure every year is displayed
plt.ylim(0.8, 1.8) 
# Show gridlines
plt.grid(True)

plt.tight_layout()  # To ensure everything fits nicely
plt.show()


# Calculate the mean
avg_fit = results_df['Fit'].mean()
avg_alpha_mle = results_df['Alp(MLE)'].mean()
avg_alpha_mad = results_df['Alp(MAD)'].mean()
avg_rmse_mle = results_df['RMSE(MLE)'].mean()
avg_rmse_mad = results_df['RMSE(MAD)'].mean()
avg_mle_err = results_df['MLE % err'].mean()
avg_mad_err = results_df['MAD % err'].mean()

# Calculate the median
median_fit = results_df['Fit'].median()
median_alpha_mle = results_df['Alp(MLE)'].median()
median_alpha_mad = results_df['Alp(MAD)'].median()
median_rmse_mle = results_df['RMSE(MLE)'].median()
median_rmse_mad = results_df['RMSE(MAD)'].median()
median_mle_err = results_df['MLE % err'].median()
median_mad_err = results_df['MAD % err'].median()

# Calculate the standard deviation
std_fit = results_df['Fit'].std()
std_alpha_mle = results_df['Alp(MLE)'].std()
std_alpha_mad = results_df['Alp(MAD)'].std()
std_rmse_mle = results_df['RMSE(MLE)'].std()
std_rmse_mad = results_df['RMSE(MAD)'].std()
std_mle_err = results_df['MLE % err'].std()
std_mad_err = results_df['MAD % err'].std()

# Print the calculated statistics
print("Mean:")
print("  Fit:", round(avg_fit, 1))
print("  Alp(MLE):", round(avg_alpha_mle, 1))
print("  Alp(MAD):", round(avg_alpha_mad, 1))
print("  RMSE(MLE):", round(avg_rmse_mle, 1))
print("  RMSE(MAD):", round(avg_rmse_mad, 1))
print("  MLE % err:", round(avg_mle_err, 1))
print("  MAD % err:", round(avg_mad_err, 1))

print("\nMedian:")
print("  Fit:", round(median_fit, 1))
print("  Alp(MLE):", round(median_alpha_mle, 1))
print("  Alp(MAD):", round(median_alpha_mad, 1))
print("  RMSE(MLE):", round(median_rmse_mle, 1))
print("  RMSE(MAD):", round(median_rmse_mad, 1))
print("  MLE % err:", round(median_mle_err, 1))
print("  MAD % err:", round(median_mad_err, 1))

print("\nStandard Deviation:")
print("  Fit:", round(std_fit, 1))
print("  Alp(MLE):", round(std_alpha_mle, 1))
print("  Alp(MAD):", round(std_alpha_mad, 1))
print("  RMSE(MLE):", round(std_rmse_mle, 1))
print("  RMSE(MAD):", round(std_rmse_mad, 1))
print("  MLE % err:", round(std_mle_err, 1))
print("  MAD % err:", round(std_mad_err, 1))







```

```python
import numpy as np
import pandas as pd

# Given alpha values
alphas = [1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.4, 1.4, 1.5, 1.4, 1.4, 1.4, 1.4, 1.4, 1.4, 1.4, 1.4, 1.3, 1.4, 1.4, 1.3, 1.3, 1.3]

# Define functions for MAD-based and Quantile-based Skewness and Kurtosis
def mad_skewness(alpha):
    return round(-1 + (np.power(2, 1/alpha) / (alpha * (np.power(2, 1/alpha) - 1))), 2)

def quantile_skewness(alpha):
    return round(-1 + (2 * (np.power(4, 1/alpha) - np.power(2, 1/alpha)) / (np.power(4, 1/alpha) - np.power(4/3, 1/alpha))), 2)

def n_m(alpha):
    return 3 * np.power(4/3, 1/alpha) + np.power(4, 1/alpha) - 2 * np.power(2, 1/alpha) - 2

def n_q(alpha):
    return (np.power(8/5, 1/alpha) - np.power(8/7, 1/alpha)) + (np.power(8, 1/alpha) - np.power(8/3, 1/alpha))

def mad_kurtosis(alpha):
    return round(n_m(alpha) / (2 * (np.power(2, 1/alpha) - 1)), 2)

def quantile_kurtosis(alpha):
    return round(n_q(alpha) / (np.power(4, 1/alpha) - np.power(4/3, 1/alpha)), 2)

# Calculate Skewness and Kurtosis
mad_skews = [mad_skewness(alpha) for alpha in alphas]
quantile_skews = [quantile_skewness(alpha) for alpha in alphas]

mad_kurtoses = [mad_kurtosis(alpha) for alpha in alphas]
quantile_kurtoses = [quantile_kurtosis(alpha) for alpha in alphas]

# Calculate the statistics
stats = {
    'Mean': [round(np.mean(alphas), 2), round(np.mean(mad_skews), 2), round(np.mean(quantile_skews), 2), 
             round(np.mean(mad_kurtoses), 2), round(np.mean(quantile_kurtoses), 2)],
    'Median': [round(np.median(alphas), 2), round(np.median(mad_skews), 2), round(np.median(quantile_skews), 2),
               round(np.median(mad_kurtoses), 2), round(np.median(quantile_kurtoses), 2)],
    'Max': [round(np.max(alphas), 2), round(np.max(mad_skews), 2), round(np.max(quantile_skews), 2), 
           round(np.max(mad_kurtoses), 2), round(np.max(quantile_kurtoses), 2)],
    'Min': [round(np.min(alphas), 2), round(np.min(mad_skews), 2), round(np.min(quantile_skews), 2),
           round(np.min(mad_kurtoses), 2), round(np.min(quantile_kurtoses), 2)]
}

# Create a DataFrame for the table
df = pd.DataFrame(stats, index=['Alpha', 'MAD Skewness', 'Quantile Skewness', 'MAD Kurtosis', 'Quantile Kurtosis'])

print("Skewness and Kurtosis Values:")
print("MAD-based Skewness:", mad_skews)
print("Quantile-based Skewness:", quantile_skews)
print("MAD-based Kurtosis:", mad_kurtoses)
print("Quantile-based Kurtosis:", quantile_kurtoses)

print("\nStatistical Summary:")
print(df)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from scipy.optimize import newton
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('Money for different gender.csv')  # Adjust the file path as needed

# Initialize the dictionary to store alpha values
alpha_values = {
    'Year': [],
    'Female Alpha (MAD)': [],
    'Male Alpha (MAD)': [],
}

# Function to calculate alpha using MAD and Newton's method
def calculate_alpha_mad(data, num_iterations):
    alpha_mad_values = []
    for _ in range(num_iterations):
        data = data.dropna()
        train_data, _ = train_test_split(data, test_size=0.5)

        # Calculate median and MAD
        median = np.median(train_data)
        mad = np.mean(np.abs(train_data - median))
        beta_MAD = np.min(train_data)

        # Define the equation for MAD
        def equation(alpha):
            return mad - (alpha * beta_MAD * (2**(1/alpha) - 1) / (alpha - 1))

        # Solve the equation for alpha using Newton's method
        alpha_MAD = newton(equation, 1.1)
        alpha_mad_values.append(alpha_MAD)

    return np.mean(alpha_mad_values)

# Get the columns for female and male data based on the year range
female_columns = [f'FemaleWorth{i}' for i in range(2010, 2024)]
male_columns = [f'MaleWorth{i}' for i in range(2010, 2024)]

# Number of iterations for each year
num_iterations = 100

# Iterate through each year and calculate alpha for females and males
for year, female_col, male_col in zip(range(2010, 2024), female_columns, male_columns):
    alpha_values['Year'].append(year)
    female_data = df[female_col]
    male_data = df[male_col]

    # Calculate alpha using MAD
    alpha_female = calculate_alpha_mad(female_data, num_iterations)
    alpha_male = calculate_alpha_mad(male_data, num_iterations)

    alpha_values['Female Alpha (MAD)'].append(alpha_female)
    alpha_values['Male Alpha (MAD)'].append(alpha_male)

# Create a DataFrame from the alpha values dictionary
results_df = pd.DataFrame(alpha_values)

# Print the results in table form
print("Alpha values using MAD for Female and Male (2010-2023):")
print(results_df)

# Plot the results with circles marking each year
plt.figure(figsize=(10, 6))
plt.plot(results_df['Year'], results_df['Female Alpha (MAD)'], label='Female α (MAD)', linestyle='-', color='k', marker='o')  # Dashed black line with circles
plt.plot(results_df['Year'], results_df['Male Alpha (MAD)'], label='Male α (MAD)', linestyle='--', color='k', marker='*')  # Solid black line with circles



# Set the axis labels and title
plt.xlabel('Year')
plt.ylabel('α')
plt.title('Pareto Alpha Values (MAD) for Female and Male (2010-2023)')

# Add a legend and a grid
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import newton
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

file_paths = [
    ("Asia_Fortune.csv", "Asia"),
    ("Europe_Fortune.csv", "Europe"),
    ("North_America_Fortune.csv", "North America")
]

def calculate_alpha_mad(data, num_iterations=100):
    alpha_mad_values = []
    for _ in range(num_iterations):
        data = data.dropna()
        train_data, _ = train_test_split(data, test_size=0.5)

        median = np.median(train_data)
        mad = np.mean(np.abs(train_data - median))
        beta_MAD = np.min(train_data)

        def equation(alpha):
            return mad - (alpha * beta_MAD * (2**(1/alpha) - 1) / (alpha - 1))

        alpha_MAD = newton(equation, 1.1)
        alpha_mad_values.append(alpha_MAD)

    return np.mean(alpha_mad_values)

alpha_dfs = []

for file_path, continent_name in file_paths:
    data = pd.read_csv(file_path)
    alpha_df = pd.DataFrame(columns=["Year", "Alpha"])

    for year in range(2001, 2024):
        year_data = data[data["Year"] == year]
        worth_data = year_data["Worth"]
        alpha_mad = calculate_alpha_mad(worth_data)
        alpha_df = pd.concat([alpha_df, pd.DataFrame({"Year": [year], "Alpha": [alpha_mad]})], ignore_index=True)

    alpha_dfs.append((alpha_df, continent_name))

# Concatenate all alpha data into one DataFrame
all_alpha_data = pd.concat([df.set_index('Year') for df, _ in alpha_dfs], axis=1, keys=[name for _, name in alpha_dfs])

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(all_alpha_data.index, all_alpha_data['Asia'], label='Asia α (MAD)', linestyle='-', color='k', marker='o')  # Solid black line with circle markers
plt.plot(all_alpha_data.index, all_alpha_data['Europe'], label='Europe α (MAD)', linestyle='--', color='k', marker='x')  # Dashed black line with x markers
plt.plot(all_alpha_data.index, all_alpha_data['North America'], label='North America α (MAD)', linestyle='-.', color='k', marker='*')  # Dash-dot black line with star markers

plt.title("Pareto Distribution Alpha (MAD) Over the Years by Continent")
plt.xlabel("Year")
plt.ylabel("α")
# Rotate x-axis labels vertically
plt.xticks(np.arange(2001, 2024), rotation='vertical')  # Rotate labels
# Add a legend and a grid
plt.legend()
plt.grid(True)

# Tight layout to prevent label overlapping
plt.tight_layout()

# Show the plot
plt.show()

# Print the table
for df, continent_name in alpha_dfs:
    print(f"\n{continent_name} Alpha (MAD) over the years:\n{df}")
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.integrate import simps


alpha = 20  
x_min = 10  


sample_size = 500
pareto_sample = stats.pareto.rvs(alpha, scale=x_min, size=sample_size)


pareto_sample_sorted = np.sort(pareto_sample)
cdf = np.arange(1, sample_size + 1) / sample_size


median = np.median(pareto_sample)


folded_cdf = np.where(pareto_sample_sorted > median, 1 - cdf, cdf)

cdf_area = simps(cdf, pareto_sample_sorted)
folded_cdf_area = simps(folded_cdf, pareto_sample_sorted)

print("Area under the CDF:", cdf_area)
print("Area under the Folded CDF:", folded_cdf_area)


plt.figure(figsize=(10, 6))
plt.plot(pareto_sample_sorted, cdf, label='CDF')
plt.plot(pareto_sample_sorted, folded_cdf, label='Folded CDF')



plt.axvline(median, color='gray', linestyle='--', label=f'Median: {median:.2f}')

plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('CDF and Folded CDF of Pareto Distribution')
plt.legend()
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
from scipy.integrate import quad


def quartiles_method_1(alpha, x_m):
    q1 = x_m * (4 / 3) ** (1 / alpha)
    q2 = x_m * 2 ** (1 / alpha)
    q3 = x_m * 4 ** (1 / alpha)
    return q1, q2, q3

def quartiles_pareto(alpha, x_m):
    q1 = x_m / (1 - 0.25) ** (1 / alpha)
    q2 = x_m / (1 - 0.5) ** (1 / alpha)
    q3 = x_m / (1 - 0.75) ** (1 / alpha)
    return q1, q2, q3


def compute_madp(alpha, x_m, p):

    m_p = x_m / (1 - p) ** (1 / alpha)

    def madp_integrand(x):
        return max((1 - p) / p * (m_p - x), x - m_p) * alpha * x_m ** alpha/(x**(alpha+1))


    madp_value, _ = quad(madp_integrand, x_m, np.inf)
    return madp_value


alpha = 1.5
x_m = 10  


q1_1, q2_1, q3_1 = quartiles_method_1(alpha, x_m)
q1_2, q2_2, q3_2 = quartiles_pareto(alpha, x_m)


mad1 = compute_madp(alpha, x_m, 0.25)  
mad2 = compute_madp(alpha, x_m, 0.5)   
mad3 = compute_madp(alpha, x_m, 0.75)  


print("Method 1 Quartiles:")
print("First Quartile:", q1_1)
print("Median:", q2_1)
print("Third Quartile:", q3_1)

print("\nPareto Method Quartiles:")
print("First Quartile:", q1_2)
print("Median:", q2_2)
print("Third Quartile:", q3_2)

print("\nMADp Values:")
print("MAD1 (p=0.25):", mad1)
print("MAD2 (p=0.5):", mad2)
print("MAD3 (p=0.75):", mad3)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.integrate import simps

# Pareto分布参数
alpha = 20
x_min = 10

# 生成样本
sample_size = 500
pareto_sample = stats.pareto.rvs(alpha, scale=x_min, size=sample_size)
pareto_sample_sorted = np.sort(pareto_sample)
cdf = np.arange(1, sample_size + 1) / sample_size

# 中位数和第一次折叠
median = np.median(pareto_sample)
folded_cdf_1 = np.where(pareto_sample_sorted > median, 1 - cdf, cdf)

# 面积
cdf_area = simps(cdf, pareto_sample_sorted)
folded_cdf_area_1 = simps(folded_cdf_1, pareto_sample_sorted)

print("Area under the CDF:", cdf_area)
print("Area under the Folded CDF (first fold):", folded_cdf_area_1)

plt.figure(figsize=(10, 6))
plt.plot(pareto_sample_sorted, cdf, label='Original CDF', color='blue')
plt.plot(pareto_sample_sorted, folded_cdf_1, label='Folded CDF', color='orange')
plt.axvline(median, color='gray', linestyle='--', label=f'Median: {median:.2f}')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('CDF and First Folded CDF of Pareto Distribution')
plt.legend()
plt.show()

# 第二次折叠
p_25 = np.percentile(pareto_sample_sorted, 25)
folded_cdf_2 = np.where(pareto_sample_sorted > p_25, 1 - folded_cdf_1, folded_cdf_1)

folded_cdf_area_2 = simps(folded_cdf_2, pareto_sample_sorted)

print("Area under the Folded CDF (second fold):", folded_cdf_area_2)

plt.figure(figsize=(10, 6))
plt.plot(pareto_sample_sorted, folded_cdf_1, label='First Folded CDF', color='orange')
plt.plot(pareto_sample_sorted, folded_cdf_2, label='Second Folded CDF', color='green')
plt.axvline(p_25, color='red', linestyle='--', label=f'25th Percentile: {p_25:.2f}')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('First and Second Folded CDFs of Pareto Distribution')
plt.legend()
plt.show()
```
